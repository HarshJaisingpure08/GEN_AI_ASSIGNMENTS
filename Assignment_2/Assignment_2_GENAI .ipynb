```python
!pip install -q cohere
import cohere
from google.colab import userdata

# Initializing our cohere here so with our api key kept safe as secret
api_key = userdata.get('COHERE_API_KEY')
co = cohere.ClientV2(api_key)

def get_response(prompt):
    response = co.chat(
        model="command-a-03-2025",
        messages=[{"role": "user", "content": prompt}]
    )
    return response.message.content[0].text
```

ZERO SHOT PROMPTING


```python
# Zero-Shot: Simple request
zero_shot_prompt = "Classify the sentiment of: 'The battery life is okay, but the screen is amazing.'"
print(" Zero-Shot Result ")
print(get_response(zero_shot_prompt))

# Few-Shot: Providing examples
few_shot_prompt = """Classify the sentiment into Positive, Negative, or Neutral.
Review: The food was cold. | Sentiment: Negative
Review: I loved the ambiance! | Sentiment: Positive
Review: The battery life is okay, but the screen is amazing. | Sentiment:"""
print("\n Few-Shot Result ")
print(get_response(few_shot_prompt))
```

```python
# 1. Interview Approach (Systematic breakdown)
interview_prompt = """I need to extract data from this email: 'Hi, I'm Jane. My order #12345 arrived broken.'
Follow these steps:
1. Identify the user.
2. Extract the order ID.
3. Identify the core issue.
4. Output as JSON."""

# 2. Chain of Thought (CoT)
cot_prompt = """Question: I have 3 apples. I buy 2 more crates. Each crate has 10 apples.
I give 5 apples to a friend. How many do I have?
Let's think step by step:"""

# 3. Tree of Thoughts (ToT) - Simplified for single prompt
tot_prompt = """Task: Design a marketing slogan for a solar-powered watch.
Phase 1: Generate 3 distinct creative directions.
Phase 2: Critique each direction for 'uniqueness' and 'clarity'.
Phase 3: Select the best one and expand it."""

print("Interview Approach")
print(get_response(interview_prompt))
print("\n Chain of Thought")
print(get_response(cot_prompt))
print("\n Tree of Thoughts --")
print(get_response(tot_prompt))
```
